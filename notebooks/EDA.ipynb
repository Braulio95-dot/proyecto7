```python
import pandas as pd

df = pd.read_csv('vehicles_us.csv')

df.head()

df.info()

df['is_4wd'] = df['is_4wd'].fillna(0).astype(int)

df['odometer'] = df['odometer'].fillna(df['odometer'].mean())


```

```python
import plotly.express as px

fig = px.scatter(df, x='model_year', y='price', color='condition',
                 title='Precio vs Año del Modelo (coloreado por condición)')
fig.show()
```

```python
fig = px.histogram(df, x='odometer', nbins=30,
                   title='Distribución del Odómetro (kilometraje)')
fig.show()
```

```python
fig = px.box(df, x='type', y='price', color='type',
             title='Distribución del Precio por Tipo de Vehículo')
fig.show()
```

```python
fig = px.violin(df, x='condition', y='days_listed', box=True, points="all",
                title='Días en lista según la condición del vehículo')
fig.show()

```

```python
fuel_counts = df['fuel'].value_counts().reset_index()
fuel_counts.columns = ['fuel_type', 'count']  # Renombramos las columnas

fig = px.bar(fuel_counts,
             x='fuel_type', y='count',
             labels={'fuel_type': 'Tipo de combustible', 'count': 'Cantidad'},
             title='Cantidad de vehículos por tipo de combustible')

fig.show()
```
